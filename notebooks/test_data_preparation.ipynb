# Data Preparation


```python
import os
from prompt2yolo.utils.s3_handler import S3Handler
from prompt2yolo.data_generation.image_generator import DreamBoothGenerator
from prompt2yolo.data_generation.image_labeler import YoloWorldLabeler

s3_handler = S3Handler(
    s3_folder="model_checkpoints/dream_booth_lora_checkpoint/",
    local_dir="./tmp"
)

test_prompt = "campfire, person"
```

### S3 Connection

```python
# Check if the local directory exists, if not, create it
if not os.path.exists(s3_handler.local_dir):
    os.makedirs(s3_handler.local_dir)

# Debug: List available files in the S3 folder to verify
available_files = s3_handler.list_files_in_folder()
print(f"Files in S3 folder '{s3_handler.s3_folder}': {available_files}")
```

### Image Generation

```python
generator = DreamBoothGenerator(s3_handler=s3_handler,  num_images=1)
# Download all files in the specified folder
downloaded_files = generator.download_lora_checkpoint()
print(f"Downloaded files: {downloaded_files}")

generator.generate(prompt=test_prompt)
```

### Image Labeling

```python
image_labeler = YoloWorldLabeler(
    custom_classes=["person"],
    s3_handler=s3_handler
)

image_labeler.label()
```
